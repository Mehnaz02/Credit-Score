```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

col_l=[]
for i in data.columns:
  if ((data[i].dtype != 'object')&(i != 'Credit_Score')):
    col_l.append(i)
```

```python
x=data[col_l]
vif_data=pd.DataFrame()
vif_data['Features']=x.columns
vif_data['vif']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
```

Model Building

```python
#spliiting the data
x=data.iloc[:,:-1]
y=data.iloc[:,-1]

#training and testing data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#Standardlizing the data

from  sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_trains=sc.fit_transform(x_train)
x_tests=sc.transform(x_test)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression

lg=LogisticRegression()
lg.fit(x_trains,y_train)
y_pred=lg.predict(x_tests)
#checking accuracy for logistic regression
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
pd.DataFrame({'Actual_':y_test,'Predict_':y_pred})
```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(x_trains,y_train)
y_pred=dt.predict(x_tests)
 # checking accuracy for Decision tree

print(accuracy_score(y_test,y_pred))

pd.DataFrame({'Actual_':y_test,'Predict_':y_pred})
```

Hyperparameter Tuning on Decision Tree

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'max_features':['log2','sqrt','auto'],
            'criterion':['entropy','gini'],
            'max_depth':[2,3,5,10,50],
            'min_samples_split':[2,3,50,100],
            'min_samples_leaf':[1,5,8,10]}

grid_obj=GridSearchCV(dt,parameters)
grid_obj=grid_obj.fit(x_trains,y_train)
dt=grid_obj.best_estimator_
dt.fit(x_trains,y_train)
y_pred=dt.predict(x_tests)
acc_dt=round(accuracy_score(y_test,y_pred)*100,2)
print('Accuracy_Decisioc',acc_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(x_trains,y_train)
y_pred=rfc.predict(x_tests)

print(accuracy_score(y_test,y_pred))

pd.DataFrame({'Actual_':y_test,'Predict_':y_pred})
```
